```python
import numpy as np
np.set_printoptions(threshold=10000,suppress=True)
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
data = pd.read_csv('./villes.csv', sep=';')
X = data.iloc[:, 1:13].values
labels = data.iloc[:, 0].values

scaled_data = StandardScaler().fit_transform(X)
pca_2 = PCA(n_components=0.9)

scaled_data = pca_2.fit_transform(scaled_data)

print(pca_2.explained_variance_ratio_.cumsum())

# il faut garder 2 axes (on met le n_components a 0.9 pour garder 90% de l'information. Grace au explained_variance_ratio_.cumsum on voit que le 2ieme chiffre est superieurs a 0.90 donc il faut bien 2 axes).
#
```

```python
import matplotlib
plt.scatter(scaled_data[:, 0], scaled_data[:, 1])
for label, x, y in zip(labels, scaled_data[:, 0], scaled_data[:, 1]):
    plt.annotate(label, xy=(x, y), xytext=(-0.2, 0.2), textcoords='offset points')
plt.show()
```

```python
data = pd.read_csv('./villes.csv', sep=';')
X = data.iloc[:, 1:13].values
labels = data.iloc[:, 0].values

scaled_data = StandardScaler().fit_transform(X)
pca_2 = PCA(n_components=0.9)

scaled_data = pca_2.fit_transform(scaled_data)

print(pca_2.explained_variance_ratio_.cumsum())

kmeans = KMeans(n_clusters=3).fit_predict(scaled_data)
clustering = AgglomerativeClustering(n_clusters=3).fit_predict(scaled_data)
clusteringWa = AgglomerativeClustering(n_clusters=3, linkage='ward').fit_predict(scaled_data)
clusteringAv = AgglomerativeClustering(n_clusters=3, linkage='average').fit_predict(scaled_data)
```

```python
colors = ['red','yellow','blue','pink']
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c= kmeans, cmap=matplotlib.colors.ListedColormap(colors))
for label, x, y in zip(labels, scaled_data[:, 0], scaled_data[:, 1]):
 plt.annotate(label, xy=(x, y), xytext=(-0.2, 0.2), textcoords='offset points')
plt.show()
```

```python
colors = ['red','yellow','blue','pink']
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c= clustering, cmap=matplotlib.colors.ListedColormap(colors))
for label, x, y in zip(labels, scaled_data[:, 0], scaled_data[:, 1]):
 plt.annotate(label, xy=(x, y), xytext=(-0.2, 0.2), textcoords='offset points')
plt.show()
```

```python
colors = ['red','yellow','blue','pink']
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c= clusteringWa, cmap=matplotlib.colors.ListedColormap(colors))
for label, x, y in zip(labels, scaled_data[:, 0], scaled_data[:, 1]):
 plt.annotate(label, xy=(x, y), xytext=(-0.2, 0.2), textcoords='offset points')
plt.show()
```

```python
colors = ['red','yellow','blue','pink']
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c= clusteringAv, cmap=matplotlib.colors.ListedColormap(colors))
for label, x, y in zip(labels, scaled_data[:, 0], scaled_data[:, 1]):
 plt.annotate(label, xy=(x, y), xytext=(-0.2, 0.2), textcoords='offset points')
plt.show()
```
